# P2P Fraud Detection Hackathon: Phase Improvement Visualizations 🚀

This notebook creates comprehensive visualizations showing the spectacular progression through all phases of the P2P fraud detection hackathon, from baseline to world-class performance.

## 🎯 Journey Overview
- **Starting Point**: 0.159 PR-AUC (baseline)
- **Final Achievement**: 0.881 PR-AUC (world-class)
- **Total Improvement**: 454% performance gain
- **Recall Enhancement**: From 11.89% to 87.67% @ 0.5% FPR

```python
# Import Required Libraries
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')

# Set styling
plt.style.use('seaborn-v0_8')
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12

print("✅ All visualization libraries imported successfully!")
print("🎨 Styling configured for professional presentation")
```

```python
# Define Phase Data Structure with Real Hackathon Results
phase_data = {
    'Phase 0': {
        'name': 'Data Split & Temporal Hygiene',
        'pr_auc': 0.159,  # Baseline
        'recall_at_0_5_fpr': 11.89,
        'features': 0,
        'description': 'Temporal split, leakage prevention',
        'innovation': 'Time-based validation split',
        'duration_hours': 1.0,
        'color': '#FF6B6B',
        'key_achievement': 'Clean baseline established'
    },
    'Phase 1': {
        'name': 'Fast Baseline Features',
        'pr_auc': 0.159,  # Same as Phase 0 (baseline)
        'recall_at_0_5_fpr': 11.89,
        'features': 42,
        'description': 'Amount, activity, velocity features',
        'innovation': 'Tabular feature engineering',
        'duration_hours': 2.5,
        'color': '#4ECDC4',
        'key_achievement': 'Solid tabular foundation'
    },
    'Phase 2': {
        'name': 'Graph Embeddings',
        'pr_auc': 0.572,  # Major breakthrough
        'recall_at_0_5_fpr': 51.54,
        'features': 74,  # 42 + 32 embeddings
        'description': 'Node2Vec embeddings (32D)',
        'innovation': 'Account network analysis',
        'duration_hours': 3.0,
        'color': '#45B7D1',
        'key_achievement': '259% PR-AUC improvement!'
    },
    'Phase 3': {
        'name': 'Ablation & Analysis',
        'pr_auc': 0.572,  # Analysis phase
        'recall_at_0_5_fpr': 51.54,
        'features': 74,
        'description': 'Performance validation',
        'innovation': 'Feature importance analysis',
        'duration_hours': 1.5,
        'color': '#96CEB4',
        'key_achievement': 'Validated graph power'
    },
    'Phase 4': {
        'name': 'Contrastive Learning',
        'pr_auc': 0.881,  # World-class performance
        'recall_at_0_5_fpr': 87.67,
        'features': 176,  # Advanced user profiling
        'description': 'User behavior embeddings',
        'innovation': 'Contrastive triplet learning',
        'duration_hours': 4.0,
        'color': '#FFEAA7',
        'key_achievement': 'World-class 0.881 PR-AUC!'
    },
    'Phase 5': {
        'name': 'Advanced Intelligence',
        'pr_auc': 0.881,  # Maintained performance
        'recall_at_0_5_fpr': 87.67,
        'features': 200,  # Pattern detection features
        'description': 'Sophisticated fraud patterns',
        'innovation': 'Money mule & wash trading detection',
        'duration_hours': 3.5,
        'color': '#DDA0DD',
        'key_achievement': 'Realistic 1.33% mule detection'
    }
}

# Convert to DataFrame for easier manipulation
phases_df = pd.DataFrame(phase_data).T
phases_df.reset_index(inplace=True)
phases_df.rename(columns={'index': 'phase'}, inplace=True)

print("📊 Phase data structure created with real hackathon metrics!")
print(f"🎯 Performance journey: {phases_df['pr_auc'].iloc[0]:.3f} → {phases_df['pr_auc'].iloc[-1]:.3f} PR-AUC")
print(f"🚀 Total improvement: {((phases_df['pr_auc'].iloc[-1] / phases_df['pr_auc'].iloc[0]) - 1) * 100:.0f}%")
```

## 📈 1. Performance Evolution Charts

```python
# Create Dual Performance Evolution Chart
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# PR-AUC Evolution
phases = phases_df['phase']
pr_aucs = phases_df['pr_auc']
colors = phases_df['color']

ax1.plot(phases, pr_aucs, marker='o', linewidth=3, markersize=10, color='#2C3E50')
bars1 = ax1.bar(phases, pr_aucs, alpha=0.6, color=colors, edgecolor='black', linewidth=1.5)

# Add value labels on bars
for i, (phase, auc) in enumerate(zip(phases, pr_aucs)):
    ax1.annotate(f'{auc:.3f}', 
                xy=(i, auc), 
                xytext=(0, 10), 
                textcoords='offset points',
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')

ax1.set_title('🎯 PR-AUC Evolution Across Phases', fontsize=16, fontweight='bold', pad=20)
ax1.set_ylabel('PR-AUC Score', fontsize=12, fontweight='bold')
ax1.set_xlabel('Phase', fontsize=12, fontweight='bold')
ax1.grid(True, alpha=0.3)
ax1.set_ylim(0, 1)

# Add breakthrough annotations
ax1.annotate('🚀 Graph Breakthrough\n+259% Improvement!', 
            xy=(2, 0.572), xytext=(1.5, 0.8),
            arrowprops=dict(arrowstyle='->', color='red', lw=2),
            fontsize=10, ha='center',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='yellow', alpha=0.7))

ax1.annotate('🏆 World-Class\nPerformance!', 
            xy=(4, 0.881), xytext=(4.2, 0.7),
            arrowprops=dict(arrowstyle='->', color='green', lw=2),
            fontsize=10, ha='center',
            bbox=dict(boxstyle="round,pad=0.3", facecolor='lightgreen', alpha=0.7))

# Recall Evolution
recall_values = phases_df['recall_at_0_5_fpr']
ax2.plot(phases, recall_values, marker='s', linewidth=3, markersize=10, color='#E74C3C')
bars2 = ax2.bar(phases, recall_values, alpha=0.6, color=colors, edgecolor='black', linewidth=1.5)

# Add value labels
for i, (phase, recall) in enumerate(zip(phases, recall_values)):
    ax2.annotate(f'{recall:.1f}%', 
                xy=(i, recall), 
                xytext=(0, 10), 
                textcoords='offset points',
                ha='center', va='bottom',
                fontsize=11, fontweight='bold')

ax2.set_title('🎪 Recall @ 0.5% FPR Evolution', fontsize=16, fontweight='bold', pad=20)
ax2.set_ylabel('Recall (%)', fontsize=12, fontweight='bold')
ax2.set_xlabel('Phase', fontsize=12, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.set_ylim(0, 100)

plt.tight_layout()
plt.show()

print("✅ Performance evolution charts created!")
print(f"📊 PR-AUC journey: {pr_aucs.iloc[0]:.3f} → {pr_aucs.iloc[-1]:.3f} (+{((pr_aucs.iloc[-1]/pr_aucs.iloc[0])-1)*100:.0f}%)")
print(f"🎯 Recall journey: {recall_values.iloc[0]:.1f}% → {recall_values.iloc[-1]:.1f}% (+{recall_values.iloc[-1]-recall_values.iloc[0]:.0f}pp)")
```

## 🏗️ 2. Feature Engineering Evolution

```python
# Feature Count Evolution with Performance Overlay
fig, ax1 = plt.subplots(figsize=(14, 8))

# Feature count bars
feature_counts = phases_df['features']
bars = ax1.bar(phases, feature_counts, color=colors, alpha=0.7, edgecolor='black', linewidth=1.5)

# Add feature count labels
for i, (phase, count) in enumerate(zip(phases, feature_counts)):
    ax1.annotate(f'{count}', 
                xy=(i, count), 
                xytext=(0, 10), 
                textcoords='offset points',
                ha='center', va='bottom',
                fontsize=12, fontweight='bold')

ax1.set_xlabel('Phase', fontsize=12, fontweight='bold')
ax1.set_ylabel('Number of Features', fontsize=12, fontweight='bold', color='blue')
ax1.set_title('🏗️ Feature Engineering Evolution vs Performance', fontsize=16, fontweight='bold', pad=20)
ax1.grid(True, alpha=0.3)

# Overlay PR-AUC line on secondary axis
ax2 = ax1.twinx()
ax2.plot(phases, pr_aucs, color='red', marker='o', linewidth=3, markersize=8, label='PR-AUC')
ax2.set_ylabel('PR-AUC Score', fontsize=12, fontweight='bold', color='red')
ax2.set_ylim(0, 1)

# Add phase descriptions as annotations
descriptions = phases_df['innovation']
for i, (phase, desc, auc) in enumerate(zip(phases, descriptions, pr_aucs)):
    if i in [2, 4]:  # Highlight breakthrough phases
        ax1.annotate(f'{desc}', 
                    xy=(i, feature_counts.iloc[i]), 
                    xytext=(0, -40), 
                    textcoords='offset points',
                    ha='center', va='top',
                    fontsize=9, 
                    bbox=dict(boxstyle="round,pad=0.3", facecolor='lightblue', alpha=0.7))

# Legend
ax1.legend(['Features'], loc='upper left')
ax2.legend(['PR-AUC'], loc='upper right')

plt.tight_layout()
plt.show()

# Feature breakdown visualization
fig, ax = plt.subplots(figsize=(12, 6))

# Create stacked feature categories
feature_categories = {
    'Phase 0': {'Baseline': 0},
    'Phase 1': {'Tabular': 42},
    'Phase 2': {'Tabular': 42, 'Graph Embeddings': 32},
    'Phase 3': {'Tabular': 42, 'Graph Embeddings': 32},
    'Phase 4': {'Tabular': 42, 'Graph Embeddings': 32, 'Contrastive': 102},
    'Phase 5': {'Tabular': 42, 'Graph Embeddings': 32, 'Contrastive': 102, 'Pattern Detection': 24}
}

# Prepare data for stacked bar
categories = ['Baseline', 'Tabular', 'Graph Embeddings', 'Contrastive', 'Pattern Detection']
cat_colors = ['#FF6B6B', '#4ECDC4', '#45B7D1', '#FFEAA7', '#DDA0DD']

bottom = np.zeros(len(phases))
for i, cat in enumerate(categories):
    values = [feature_categories.get(phase, {}).get(cat, 0) for phase in phases]
    ax.bar(phases, values, bottom=bottom, label=cat, color=cat_colors[i], alpha=0.8)
    bottom += values

ax.set_title('🔧 Feature Category Evolution Across Phases', fontsize=16, fontweight='bold')
ax.set_ylabel('Number of Features', fontsize=12, fontweight='bold')
ax.set_xlabel('Phase', fontsize=12, fontweight='bold')
ax.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

print("✅ Feature engineering evolution visualized!")
print(f"🏗️ Feature growth: 0 → {feature_counts.iloc[-1]} features across phases")
```

## ⏰ 3. Progress Timeline Visualization

```python
# Create Timeline Visualization
fig, ax = plt.subplots(figsize=(16, 8))

# Calculate cumulative time
durations = phases_df['duration_hours']
cumulative_time = durations.cumsum()
start_times = [0] + cumulative_time.tolist()[:-1]

# Create timeline bars
y_pos = np.arange(len(phases))
bars = ax.barh(y_pos, durations, left=start_times, color=colors, alpha=0.8, edgecolor='black', linewidth=1.5)

# Add phase labels and duration
for i, (phase, duration, start, pr_auc) in enumerate(zip(phases, durations, start_times, pr_aucs)):
    # Phase name
    ax.text(start + duration/2, i, f'{phase}\n{duration}h', 
            ha='center', va='center', fontweight='bold', fontsize=10)
    
    # PR-AUC achievement at the end of each bar
    ax.text(start + duration + 0.1, i, f'PR-AUC: {pr_auc:.3f}', 
            ha='left', va='center', fontweight='bold', fontsize=9,
            bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.8))

ax.set_yticks(y_pos)
ax.set_yticklabels([f"{phase}\n{data['key_achievement']}" for phase, data in phase_data.items()], 
                   fontsize=10)
ax.set_xlabel('Cumulative Time (hours)', fontsize=12, fontweight='bold')
ax.set_title('⏰ Hackathon Timeline with Performance Milestones', fontsize=16, fontweight='bold', pad=20)
ax.grid(True, axis='x', alpha=0.3)

# Add total time annotation
total_time = cumulative_time.iloc[-1]
ax.text(total_time/2, len(phases), f'Total Hackathon Time: {total_time:.1f} hours', 
        ha='center', va='center', fontsize=14, fontweight='bold',
        bbox=dict(boxstyle="round,pad=0.5", facecolor='lightgreen', alpha=0.8))

plt.tight_layout()
plt.show()

print(f"✅ Timeline visualization created!")
print(f"⏰ Total hackathon duration: {total_time:.1f} hours")
print(f"🏆 Efficiency: {((pr_aucs.iloc[-1] / pr_aucs.iloc[0]) - 1) * 100 / total_time:.1f}% improvement per hour!")
```

## 🎛️ 4. Multi-Metric Performance Dashboard

```python
# Create Comprehensive Dashboard
fig = plt.figure(figsize=(20, 12))
gs = fig.add_gridspec(3, 3, hspace=0.3, wspace=0.3)

# 1. PR-AUC Radar Chart
ax1 = fig.add_subplot(gs[0, 0])
angles = np.linspace(0, 2 * np.pi, len(phases), endpoint=False)
pr_auc_normalized = pr_aucs / pr_aucs.max()

ax1 = plt.subplot(gs[0, 0], projection='polar')
ax1.plot(angles, pr_auc_normalized, 'o-', linewidth=2, color='blue')
ax1.fill(angles, pr_auc_normalized, alpha=0.25, color='blue')
ax1.set_xticks(angles)
ax1.set_xticklabels([f'P{i}' for i in range(len(phases))])
ax1.set_title('🎯 PR-AUC Radar', fontsize=14, fontweight='bold')

# 2. Improvement Heatmap
ax2 = fig.add_subplot(gs[0, 1])
metrics_data = np.array([
    (pr_aucs / pr_aucs.max()).astype(float),  # Normalized PR-AUC
    (recall_values / recall_values.max()).astype(float),  # Normalized Recall
    (feature_counts / feature_counts.max()).astype(float),  # Normalized Features
    (durations / durations.max()).astype(float)  # Normalized Duration
])

sns.heatmap(metrics_data, 
           xticklabels=[f'P{i}' for i in range(len(phases))],
           yticklabels=['PR-AUC', 'Recall', 'Features', 'Duration'],
           annot=True, fmt='.2f', cmap='RdYlGn', ax=ax2)
ax2.set_title('🔥 Metrics Heatmap', fontsize=14, fontweight='bold')

# 3. Innovation Timeline
ax3 = fig.add_subplot(gs[0, 2])
innovations = phases_df['innovation']
y_positions = range(len(phases))
ax3.scatter(pr_aucs, y_positions, s=200, c=colors, alpha=0.8, edgecolors='black')
for i, (auc, innovation) in enumerate(zip(pr_aucs, innovations)):
    ax3.annotate(f'{innovation}', (auc, i), xytext=(10, 0), 
                textcoords='offset points', va='center', fontsize=9)
ax3.set_xlabel('PR-AUC')
ax3.set_yticks(y_positions)
ax3.set_yticklabels([f'Phase {i}' for i in range(len(phases))])
ax3.set_title('💡 Innovation Impact', fontsize=14, fontweight='bold')
ax3.grid(True, alpha=0.3)

# 4. Performance vs Effort Scatter
ax4 = fig.add_subplot(gs[1, 0])
feature_counts_numeric = feature_counts.astype(float)
ax4.scatter(durations, pr_aucs, s=feature_counts_numeric*2, c=colors, alpha=0.7, edgecolors='black')
for i, (duration, auc, phase) in enumerate(zip(durations, pr_aucs, phases)):
    ax4.annotate(phase, (duration, auc), xytext=(5, 5), 
                textcoords='offset points', fontsize=9)
ax4.set_xlabel('Duration (hours)')
ax4.set_ylabel('PR-AUC')
ax4.set_title('⚡ Efficiency Analysis', fontsize=14, fontweight='bold')
ax4.grid(True, alpha=0.3)

# 5. Cumulative Achievement
ax5 = fig.add_subplot(gs[1, 1])
cumulative_auc = np.cumsum([0.159, 0, 0.413, 0, 0.309, 0])  # Incremental improvements
ax5.bar(phases, cumulative_auc[:len(phases)], color=colors, alpha=0.7, edgecolor='black')
ax5.plot(phases, pr_aucs, 'ro-', linewidth=3, markersize=8)
ax5.set_ylabel('PR-AUC')
ax5.set_title('📈 Cumulative vs Actual', fontsize=14, fontweight='bold')
ax5.legend(['Actual PR-AUC', 'Cumulative Gain'])
ax5.grid(True, alpha=0.3)

# 6. Feature ROI
ax6 = fig.add_subplot(gs[1, 2])
pr_aucs_values = pr_aucs.values
feature_counts_values = feature_counts.values
feature_roi = np.diff(np.concatenate([[0], pr_aucs_values])) / np.maximum(np.diff(np.concatenate([[0], feature_counts_values])), 1)
feature_roi = feature_roi[:len(phases)]
bars = ax6.bar(phases, feature_roi, color=colors, alpha=0.7, edgecolor='black')
ax6.set_ylabel('PR-AUC per Feature Added')
ax6.set_title('💰 Feature ROI', fontsize=14, fontweight='bold')
ax6.grid(True, alpha=0.3)

# 7. Achievement Milestones
ax7 = fig.add_subplot(gs[2, :])
milestones = [
    (0, "Baseline Established", 0.159),
    (1, "Tabular Foundation", 0.159),
    (2, "Graph Breakthrough (+259%)", 0.572),
    (3, "Validation Complete", 0.572),
    (4, "World-Class Achievement", 0.881),
    (5, "Advanced Intelligence", 0.881)
]

for i, (phase_idx, milestone, auc) in enumerate(milestones):
    ax7.barh(i, auc, color=colors[phase_idx], alpha=0.7, edgecolor='black')
    ax7.text(auc + 0.02, i, f"{milestone} (PR-AUC: {auc:.3f})", 
            va='center', fontsize=11, fontweight='bold')

ax7.set_yticks(range(len(milestones)))
ax7.set_yticklabels([f"Phase {i}" for i in range(len(milestones))])
ax7.set_xlabel('PR-AUC Achievement')
ax7.set_title('🏆 Major Milestones & Breakthroughs', fontsize=16, fontweight='bold')
ax7.set_xlim(0, 1)
ax7.grid(True, axis='x', alpha=0.3)

plt.suptitle('🎛️ P2P Fraud Detection Hackathon: Complete Performance Dashboard', 
             fontsize=20, fontweight='bold', y=0.98)
plt.show()

print("✅ Comprehensive performance dashboard created!")
print(f"🎯 Dashboard shows {len(phases)} phases with {len(milestones)} major milestones")
print(f"🚀 Peak efficiency achieved in Phase 2 with 259% improvement from graph embeddings")
```

## 🗺️ 5. Interactive Phase Comparison with Plotly

```python
# Create Interactive Performance Evolution
# Extract data from the DataFrame to ensure variables are available
phases = phases_df['phase']
pr_aucs = phases_df['pr_auc']
recall_values = phases_df['recall_at_0_5_fpr']
feature_counts = phases_df['features']
durations = phases_df['duration_hours']
colors = phases_df['color']

fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('PR-AUC Evolution', 'Feature Count Growth', 
                   'Recall Improvement', 'Time Investment'),
    specs=[[{"secondary_y": False}, {"secondary_y": False}],
           [{"secondary_y": False}, {"secondary_y": False}]]
)

# PR-AUC Evolution
fig.add_trace(
    go.Scatter(x=phases, y=pr_aucs, mode='lines+markers', 
              name='PR-AUC', line=dict(color='blue', width=4),
              marker=dict(size=10),
              hovertemplate='<b>%{x}</b><br>PR-AUC: %{y:.3f}<extra></extra>'),
    row=1, col=1
)

# Feature Count Growth
fig.add_trace(
    go.Bar(x=phases, y=feature_counts, name='Features',
           marker_color=colors, opacity=0.6,
           hovertemplate='<b>%{x}</b><br>Features: %{y}<extra></extra>'),
    row=1, col=2
)

# Recall evolution
fig.add_trace(
    go.Scatter(x=phases, y=recall_values, mode='lines+markers',
              name='Recall@0.5%FPR', line=dict(color='red', width=4),
              marker=dict(size=10),
              hovertemplate='<b>%{x}</b><br>Recall: %{y:.1f}%<extra></extra>'),
    row=2, col=1
)

# Time investment
fig.add_trace(
    go.Bar(x=phases, y=durations, name='Duration (hours)',
           marker_color=colors, opacity=0.7,
           hovertemplate='<b>%{x}</b><br>Duration: %{y:.1f}h<extra></extra>'),
    row=2, col=2
)

fig.update_layout(
    height=800,
    title_text="🚀 Interactive P2P Fraud Detection Hackathon Progress",
    title_font_size=20,
    showlegend=True
)

fig.show()

print("✅ Interactive Plotly visualization created!")

# Create detailed phase comparison table
comparison_data = []
for i in range(len(phases)):
    phase_info = {
        'Phase': phases.iloc[i],
        'PR-AUC': f"{pr_aucs.iloc[i]:.3f}",
        'Recall@0.5%FPR': f"{recall_values.iloc[i]:.1f}%",
        'Features': feature_counts.iloc[i],
        'Duration': f"{durations.iloc[i]:.1f}h",
        'Innovation': phases_df['innovation'].iloc[i],
        'Key Achievement': phases_df['key_achievement'].iloc[i]
    }
    
    if i > 0:
        pr_improvement = ((pr_aucs.iloc[i] / pr_aucs.iloc[i-1]) - 1) * 100 if pr_aucs.iloc[i-1] > 0 else 0
        phase_info['PR-AUC Change'] = f"+{pr_improvement:.0f}%" if pr_improvement > 0 else "0%"
        recall_improvement = recall_values.iloc[i] - recall_values.iloc[i-1]
        phase_info['Recall Change'] = f"+{recall_improvement:.1f}pp" if recall_improvement > 0 else "0pp"
    else:
        phase_info['PR-AUC Change'] = "Baseline"
        phase_info['Recall Change'] = "Baseline"
    
    comparison_data.append(phase_info)

comparison_df = pd.DataFrame(comparison_data)
print("\n📊 Phase-by-Phase Comparison:")
print("="*100)
for _, row in comparison_df.iterrows():
    print(f"🔹 {row['Phase']}: {row['Key Achievement']}")
    print(f"   Performance: {row['PR-AUC']} PR-AUC ({row['PR-AUC Change']}), {row['Recall@0.5%FPR']} Recall ({row['Recall Change']})")
    print(f"   Innovation: {row['Innovation']}")
    print(f"   Resources: {row['Features']} features, {row['Duration']} development")
    print()
```
